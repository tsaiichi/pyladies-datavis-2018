```python
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set(style='ticks')

rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -0.5 * x + rs.normal(size=1000)

sns.jointplot(x, y, kind='hex', stat_func=None, color='#4CB391')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(style='ticks')

df = pd.read_csv('meteorite-landings.csv')

df = df[df['reclong']<=180]

g = sns.jointplot('reclong', 'reclat', data=df,
              kind='scatter', stat_func=None, color='#4CB391')
g.fig.set_figwidth(10)
g.fig.set_figheight(5)

g.ax_joint.cla()  # clear axis
plt.sca(g.ax_joint)  # switch to the axes of g.ax_joint

x = df['reclong']
y = df['reclat']
mass = df['mass']
areas = (mass-min(mass)) / (max(mass)-min(mass))
colors, levels = pd.factorize(df['fall'])

plt.scatter(x, y, s=np.sqrt(areas)*1000, c=colors, alpha=0.3, cmap=plt.cm.PRGn)
```
