# 畫圓

```python
## jupyter notebook 要能顯示圖片所要用的
%matplotlib inline  

## 設定 matplotlib
import matplotlib.style
import matplotlib as mpl
mpl.style.use('default')
import matplotlib.pyplot as plt


## 設定畫紙
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.set_ylim(ymin=0, ymax=2)
ax.set_xlim(xmin=0, xmax=4)

## 設定要畫的圓
c = plt.Circle(
    (2, 1),  # (x, y) 位置
    0.2,  # 半徑
    lw=1,  # line weight: 線條寬度
    fill=False,  # 要不要填色
    color=(0, 0, 0, 1)  # 顏色 (r, g, b, a)
)

## 在畫紙上畫圓
ax.add_artist(c)
```

# 畫複數個圓

```python
## jupyter notebook 要能顯示圖片所要用的
%matplotlib inline  

## 設定 matplotlib
import matplotlib.style
import matplotlib as mpl
mpl.style.use('default')
import matplotlib.pyplot as plt

## 設定畫紙
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.set_ylim(ymin=0, ymax=2)
ax.set_xlim(xmin=0, xmax=4)

## 可以傳入 x, y 位置的參數
def draw_circle(x, y, circles=[]):
    c = plt.Circle(
          (x, y),  # (x, y) 位置
          0.2,  # 半徑
          lw=1,  # line weight: 線條寬度
          fill=False,  # 要不要填色
          color=(0, 0, 0, 1)  # 顏色 (r, g, b, a)
      )
    circles.append(c)


## 給 3 個不同位置的圓，存到 circle_list 中
circle_list = []
draw_circle(2-0.2, 1, circle_list)
draw_circle(2, 1, circle_list)
draw_circle(2+0.2, 1, circle_list)


## 在畫紙上畫圓
for c in circle_list:
    ax.add_artist(c)
```

```python
## jupyter notebook 要能顯示圖片所要用的
%matplotlib inline  

## 設定 matplotlib
import matplotlib.style
import matplotlib as mpl
mpl.style.use('default')
import matplotlib.pyplot as plt

## 數值相關 package
import numpy as np

## 設定畫紙
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.set_ylim(ymin=0, ymax=2)
ax.set_xlim(xmin=0, xmax=4)

r = 0.2  # 設定半徑
h = np.sqrt(3)*(r/2) # 行距

## 可以傳入 x, y 位置的參數
def draw_circle(x, y, circles=[]):
    c = plt.Circle(
          (x, y),  # (x, y) 位置
          r,  # 半徑
          lw=0.3,  # line weight: 線條寬度
          fill=False,  # 要不要填色
          color=(0, 0, 0, 1)  # 顏色 (r, g, b, a)
      )
    circles.append(c)


## 隔行不同位置
circle_list = []
for x in np.arange(0+r/2, 4, r):
    for y in np.arange(0, 2+h, 2*h):
        draw_circle(x, y, circle_list)

for x in np.arange(0, 4, r):
    for y in np.arange(0-h, 2+h, 2*h):
        draw_circle(x, y, circle_list)

    
## 大圈
c = plt.Circle((2, 5*h), 3*r, lw=1,
      fill=False, color=(0, 0, 0, 1))
circle_list.append(c)

## 在畫紙上畫圓
for c in circle_list:
    ax.add_artist(c)
```

```python
## jupyter notebook 要能顯示圖片所要用的
%matplotlib inline  

## 設定 matplotlib
import matplotlib.style
import matplotlib as mpl
mpl.style.use('default')
import matplotlib.pyplot as plt

## 數值相關 package
import numpy as np
import random

## 設定畫紙
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
ax.set_ylim(ymin=0, ymax=2)
ax.set_xlim(xmin=0, xmax=4)

## 設定半徑, 行距, x y 起終點
w = 0.2
h = np.sqrt(3)*(w/2)
outer_w = 3*w
start = 0-w
y_start = 0-h
end = 4+w
y_end = 2+h

## 可以傳入 x, y 位置的參數
def draw_circle(x, y, circles):
    r = 1 
    g = round(random.uniform(0.1, 1.0), 10)
    b = round(random.uniform(0.1, 1.0), 10)
    c = plt.Circle((x, y), w, lw=1, ls='solid', 
            ec='white', fill=True, fc=(r, g, b, 0.1),
            clip_on=False)
    circles.append(c)

## 隔行不同位置
circles = []
for x in np.arange(start+w/2, end, w):
    for y in np.arange(y_start+h, y_end, 2*h):
        draw_circle(x, y, circles)

for x in np.arange(start, end, w):
    for y in np.arange(y_start, y_end, 2*h):
        draw_circle(x, y, circles)

## 大白圓底
for x in range(1, 4):
    c = plt.Circle((x, y_start+6*h), outer_w, lw=0,
        fill=True, color=(1, 1, 1, 0.5))
    circles.append(c)

## 在畫紙上畫圓
for c in circles:
    ax.add_artist(c)

fig.savefig('graph.png')
```
